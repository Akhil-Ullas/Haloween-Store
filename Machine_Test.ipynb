```python
df_train['Year']=df_train['instock_date'].dt.year
df_train['Month']=df_train['instock_date'].dt.month
df_train['Day']=df_train['instock_date'].dt.day
df_train.head()
```

**Model Split**

```python
x=df_train.drop(columns=['Selling_Price'])
x.head()
```

```python
y=df_train['Selling_Price']
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=13)
```

**Model Selection**

```python
LR_score=cross_val_score(LinearRegression(),X=x,y=y,cv=5)
DT_score=cross_val_score(DecisionTreeRegressor(),X=x,y=y,cv=5)
RF_score=cross_val_score(RandomForestRegressor(),X=x,y=y,cv=5)
KNN_score=cross_val_score(KNeighborsRegressor(),X=x,y=y,cv=5)
AD_score=cross_val_score(AdaBoostRegressor(),X=x,y=y,cv=5)
GB_score=cross_val_score(GradientBoostingRegressor(),X=x,y=y,cv=5)
XG_score=cross_val_score(XGBRegressor(),X=x,y=y,cv=5)
svm_score=cross_val_score(SVR(),X=x,y=y,cv=5)
```

```python
print(f'Linear Regression : {np.round(np.mean(LR_score)*100,2)}%')
print(f'Decision Tree : {np.round(np.mean(DT_score)*100,2)}%')
print(f'Random Forest : {np.round(np.mean(RF_score)*100,2)}%')
print(f'KNN : {np.round(np.mean(KNN_score)*100,2)}%')
print(f'Adaboost : {np.round(np.mean(AD_score)*100,2)}%')
print(f'Gradient : {np.round(np.mean(GB_score)*100,2)}%')
print(f'XGBoost : {np.round(np.mean(XG_score)*100,2)}%')
print(f'SVM : {np.round(np.mean(svm_score)*100,2)}%')
```

```python
mean_LR_score=np.mean(LR_score)
mean_DT_score=np.mean(DT_score)
mean_RF_score=np.mean(RF_score)
mean_KNN_score=np.mean(KNN_score)
mean_AD_score=np.mean(AD_score)
mean_GB_score=np.mean(GB_score)
mean_XG_score=np.mean(XG_score)
mean_svm_score=np.mean(svm_score)

Mean_Accuracy=[mean_LR_score,mean_DT_score,mean_RF_score,mean_KNN_score,mean_AD_score,mean_GB_score,mean_XG_score,mean_svm_score]
Models=['Linear Regression','Decision Tree','Random Forest','KNN','Adaboost','GradientBoost','XGBoost','SVM']

plt.figure(figsize=(12,6))
plt.bar(Models, Mean_Accuracy, color='Purple')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy of Models')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

from sklearn.model_selection import train_test_split,KFold,StratifiedKFold,cross_val_score,GridSearchCV
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler,MinMaxScaler,LabelEncoder
from sklearn import preprocessing

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df_train=pd.read_csv('/content/drive/MyDrive/Camerin DSML/Machine Test/Airoplane Accident/train.csv')
df_train.head()
```

```python
df_train=df_train.drop(columns='Accident_ID')
```

```python
x=df_train.drop(columns='Severity')
x.head()
```

```python
y=df_train['Severity']
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
LR_score=cross_val_score(LogisticRegression(),X=x,y=y,cv=5)
DT_score=cross_val_score(DecisionTreeClassifier(),X=x,y=y,cv=5)
KN_score=cross_val_score(KNeighborsClassifier(),X=x,y=y,cv=5)
RF_score=cross_val_score(RandomForestClassifier(),X=x,y=y,cv=5)
Ada_score=cross_val_score(AdaBoostClassifier(),X=x,y=y,cv=5)
GB_score=cross_val_score(GradientBoostingClassifier(),X=x,y=y,cv=5)
XG_score=cross_val_score(XGBClassifier(),X=x,y=y,cv=5)
svm_score=cross_val_score(SVC(),X=x,y=y,cv=5)
NB_score=cross_val_score(GaussianNB(),X=x,y=y,cv=5)
```

```python
print(f'Logistic Regression : {np.round(np.mean(LR_score)*100,2)}%')
print(f'Decision Tree : {np.round(np.mean(DT_score)*100,2)}%')
print(f'KNN : {np.round(np.mean(KN_score)*100,2)}%')
print(f'Random Forest : {np.round(np.mean(RF_score)*100,2)}%')
print(f'AdaBoost : {np.round(np.mean(Ada_score)*100,2)}%')
print(f'Gradient Boost : {np.round(np.mean(GB_score)*100,2)}%')
print(f'XG Boost : {np.round(np.mean(XG_score)*100,2)}%')
print(f'SVM : {np.round(np.mean(svm_score)*100,2)}%')
print(f'Naive Bayes : {np.round(np.mean(NB_score)*100,2)}%')
```

```python
gb=GradientBoostingClassifier()
gb_params=[{'max_depth':list(range(2,7))}]
```

```python
clf=GridSearchCV(gb,gb_params,cv=5,scoring='accuracy')
clf.fit(x_train,y_train)
print(clf.best_params_)
print(clf.best_score_)
```

```python
model=GradientBoostingClassifier(max_depth=6)
model.fit(x_train,y_train)

pred_y=model.predict(x_test)
pred_y
```

```python
model.score(x_test,y_test)
```

```python
cm=confusion_matrix(y_test,pred_y)
sn.heatmap(cm,cmap='BuPu',annot=True)
plt.show()
```

```python
classification_report(y_test,pred_y)
```

4. Find the fibonacci sequence

```python
a=0
b=1
print(a,b,end=' ')
for i in range(5):
  c=a+b
  print(c,end=' ')
  a=b
  b=c
```

5. Calculate the sum of all digits in a number

```python
a=int(input('Enter a number:'))
temp=a
sum=0
rem=0
while a>0:
  rem=a%10
  sum=sum+rem
  a=a//10
print(f'Sum of digits of {temp} is {sum}')
```

7. calculate the factorial of a number
